# Linear Regression

```python
x = [1,2,3,4,5,6,7,8]
y = [2,5,1,7,9,3,6,10]

x = np.array(x)
y = np.array(y)

# Step 1: Ubah ke bentuk design matrix
A = np.vstack((x, np.ones(len(x)))).T
print(A)

# Step 2: Ubah y jadi matrix kolom
y = y[:, np.newaxis]
# y = 1D -> 2D
# Y = (8,) -> (8,1)
print(y)

# Return m sm b
# Step 3: Masukin ke rumus
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T),y)
print(alpha)
```

```python
#PseudoInverse
# alpha = A+ * y
pinv = np.linalg.pinv(A) # -> (X^TX)^-1
alpha = np.dot(pinv, y)
print(pinv)
print(alpha)

# y = mx + b
# m = alpha[0]
# n = alpha[1]

plt.plot(x, y, "g+--")
plt.plot(x, alpha[0]*x + alpha[1], "b")
plt.show()
```

# Linear Interpolation

```python
from scipy.interpolate import interp1d

x = [1,2,3,4,5]
y = [1,3,2,8,6]

x = np.array(x)
y = np.array(y)

f = interp1d(x, y)

x_new = np.linspace(1,5, 100)
y_new = f(x_new)

plt.plot(x, y, 'ro')
plt.plot(x_new, y_new, '--b')
plt.show()
```

# Cubic Spline Interpolation
menggunakan pangkat 3 untuk masing - masing interval
$$a + b(x - x_0) + c(x - x_0)^2 + d(x - x_0)^3$$

```python
from scipy.interpolate import CubicSpline

x = [1,2,3,4,5]
y = [1,3,2,8,6]

x = np.array(x)
y = np.array(y)

# Suruh Libary cari koef polinomial

f = CubicSpline(x, y)

x_new = np.linspace(1, 5, 100)
y_new = f(x_new)

plt.plot(x, y, "go")
plt.plot(x_new, y_new, "--b")
plt.show()
```

# Newton Polynomial

```python
x = [1,2,3,4,5]
y = [1,3,2,8,6]

x = np.array(x)
y = np.array(y)

plt.figure(figsize=(12,8))

# Order 1 - 6
for i in range(1, 7):
  plt.subplot(2, 3, i) # Mau disusun berapa baris dan berapa kolom
  plt.plot(x, y, "go")

  # Mencari koefisien untuk masing - masing order
  coef = np.polyfit(x, y, i)
  
  # Masukin koefisiennya ke polinomial
  y_new = np.polyval(coef, x)
  plt.plot(x, y_new)
  plt.title(f'Polinomial Order {i}')

plt.show()

```
